### **Encoding**

```python
catdata.head()
```

```python
catdata.Churn.value_counts()
```

```python
# Converting categorical values to numerical values for the 'Churn' column
# Using Manual label encoding to encode class labels to numerical values
catdata.Churn = catdata.Churn.map({' False.':0,' True.':1})
catdata.Churn
```

```python
catdata.Churn.value_counts()
```

```python
# Manual Label encoding for 'VMail_Plan' column
catdata.VMail_Plan = catdata.VMail_Plan.map({' yes':1,' no': 0 })
catdata.VMail_Plan
```

```python
catdata.VMail_Plan.value_counts()
```

```python
# Manual Label encoding for 'International Plan' column
catdata.International_Plan.value_counts()
```

```python
catdata.International_Plan.unique()
catdata.International_Plan = catdata.International_Plan.map({' no':0,' yes':1})
catdata.International_Plan
```
